```python
#parsing of Matthew in order to make a proof of concept for a graded level reader in this book
import pandas as pd
import numpy as np
from re import search 
```

```python
#Get data set into the python
mattToJohn = pd.read_excel('MattToJohn.xlsx')
gospels = np.array(mattToJohn)
```

```python
#Contains the (verse: conjugation: count) to be used for weights using the gospels data
matt = []
mark = []
luke = []
john = []
sypGospels = []
for row in gospels:
    if (search("41_Mat", str(row[0]))):
        matt.append((row[0].strip(), row[2].strip(),row[6].strip()))
```

```python
headers = ["41_Mat", "42_Mrk", "43_Luk", "44_Jhn"]

verses = []
for head in headers:
    for row in gospels:
        if (search(head, str(row[0]))):
            verses.append((row[0].strip(), row[2].strip(),row[6].strip()))
        
print(verses)
```
